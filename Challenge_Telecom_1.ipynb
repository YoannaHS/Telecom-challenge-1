# **Importando Datos**

```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/refs/heads/main/Dados/Telco-Customer-Churn.json"
```

```python
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
df
```

# **Tratamiento de Datos**

```python
# Flatten nested columns
df_flat = pd.json_normalize(data)
df_flat
```

```python
df_flat.info()
```

```python
# Convert 'account.Charges.Total' to numeric, coercing errors to NaN
df_flat['account.Charges.Total'] = pd.to_numeric(df_flat['account.Charges.Total'], errors='coerce')

# Verify the data type after conversion
print(df_flat['account.Charges.Total'].dtype)

# Check for any remaining NaN values
print("Number of NaN values in 'account.Charges.Total':", df_flat['account.Charges.Total'].isnull().sum())
```

```python
df_cleaned = df_flat.dropna(subset=['account.Charges.Total']).copy()
print("Number of rows after removing NaN values in 'account.Charges.Total':", len(df_cleaned))
```

```python
# Verificamos si hay valores únicos en cada columna
for col in df_flat.columns:
  print(f"valores únicos en la columna'{col}': {df_flat[col].nunique()}")
  if df_flat[col].nunique() < 50: # Imprima los valores solamente si no hay muchos
    print(df_flat[col].unique())
    print('-' * 50)  # Línea divisoria para mejor visualización
```

```python
 #Duplicados
print("Numero de duplicados", df_flat.duplicated().sum())
```

```python
# Check for empty strings in the DataFrame
print("Number of empty values in df_cleaned:\n", df_cleaned.eq('').sum())
```

```python
df_cleaned = df_cleaned[df_cleaned['Churn'] != ''].copy()
print("Number of rows after removing empty 'Churn' values:", len(df_cleaned))
```

```python
# Check for empty strings in the DataFrame
print("Number of empty values in df_cleaned:\n", df_cleaned.eq('').sum())
```

# Creación de visuales

**En este paso, el objetivo es comprender cómo está distribuida la variable "churn" (evasión) entre los clientes. Utiliza gráficos para visualizar la proporción de clientes que permanecieron y los que se dieron de baja.**

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Define a color map for the Churn column
color_map = {'Yes': 'salmon', 'No': 'lightblue'}

fig = px.histogram(df_cleaned, x='Churn', title='Distribución de Churn', color='Churn', color_discrete_map=color_map)
fig.update_layout(
    autosize=False,
    width=600,
    height=400
)
fig.show()
```

Ahora, exploraremos cómo se distribuye la evasión según variables categóricas, como género, tipo de contrato, método de pago, entre otras.

Este análisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudará a orientar acciones estratégicas.

```python
categorical_cols = ['customer.gender', 'account.Contract', 'account.PaymentMethod',
                    'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService']

color_map = {'Yes': 'salmon', 'No': 'lightblue'}

for col in categorical_cols:
    fig = px.histogram(df_cleaned, x=col, color='Churn', barmode='group',
                       title=f'Distribución de Churn por {col}',
                       color_discrete_map=color_map,
                       text_auto=True) # Add text labels on bars
    fig.update_layout(
        autosize=False,
        width=700,
        height=400
    )
    fig.show()
```

En este paso, explora cómo las variables numéricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasión) y los que no cancelaron.

Este análisis ayuda a entender si ciertos valores numéricos están más asociados con la evasión, proporcionando insights sobre los factores que influyen en el comportamiento de los clientes.

```python
# Calculate churn rate by tenure
churn_by_tenure = df_cleaned.groupby('customer.tenure')['Churn'].value_counts(normalize=True).unstack().fillna(0)
churn_rate_by_tenure = churn_by_tenure['Yes']

# Create a line plot of churn rate by tenure
fig = px.line(x=churn_rate_by_tenure.index, y=churn_rate_by_tenure.values,
              title='Churn Rate by Tenure',
              labels={'x': 'Tenure (Months)', 'y': 'Churn Rate'})

fig.update_layout(
    autosize=False,
    width=800,
    height=500
)

fig.show()
```

```python
# Calculate the average Monthly Charges for churned and non-churned customers
average_charges_by_churn = df_cleaned.groupby('Churn')['account.Charges.Monthly'].mean().reset_index()

# Create a bar plot
fig = px.bar(average_charges_by_churn, x='Churn', y='account.Charges.Monthly',
             title='Average Monthly Charges by Churn Status',
             color='Churn', color_discrete_map={'Yes': 'salmon', 'No': 'lightblue'},
             text_auto=True) # Add text labels on bars

fig.update_layout(
    autosize=False,
    width=600,
    height=400
)

fig.show()
```

```python
 # Agrupar por tenure y Churn
df_grouped = df_cleaned.groupby(['customer.tenure', 'Churn']).size().unstack()

 # Calcular el % de churn por tenure
df_grouped['churn_rate'] = df_grouped['Yes'] / (df_grouped['Yes'] + df_grouped['No']) * 100

 # Mostrar el resultado
print(df_grouped)
```

```python
fig = px.line(df_grouped.reset_index(),
              x='customer.tenure',
              y='churn_rate',
              markers=True,
              title='Tasa de Churn (%) por Tenure')
fig.update_layout(
    autosize=False,
    width=1200,
    height=600
)
fig.show()
```

# Recomendaciones finales

# 📋 Comprendiendo y Abordando la Evasión de Clientes

Nuestro análisis de datos de clientes revela patrones críticos que influyen en la evasión. Abordar estos factores de forma estratégica puede tener un impacto significativo en la retención de clientes y el crecimiento del negocio.

---

## 🔍 Hallazgos Clave y Recomendaciones

### 1. La Evasión se Concentra en Segmentos Específicos
**Observación:** Algunos grupos de clientes, especialmente aquellos con contratos mes a mes, que utilizan pagos por cheque electrónico y que tienen servicio de internet por fibra óptica, presentan tasas de evasión significativamente más altas.

**Recomendación:**
- Prioriza los esfuerzos de retención en estos segmentos de alto riesgo.
- Desarrolla campañas específicas, programas de fidelización o mejoras en el servicio para estos grupos.
- Investiga las causas raíz dentro de estos segmentos (por ejemplo, problemas de calidad del servicio de fibra óptica, fricciones con el método de pago) para implementar soluciones duraderas.

---

### 2. La Antigüedad Temprana del Cliente es Crítica
**Observación:** Una parte sustancial de la evasión ocurre en los primeros meses del contrato del cliente. La tasa de evasión disminuye notablemente a medida que los clientes permanecen más tiempo.

**Recomendación:**
- Mejora la experiencia de incorporación de nuevos clientes.
- Implementa soporte proactivo, recursos educativos y programas de compromiso temprano para generar valor y reducir la deserción inicial.
- Considera incentivos para que los clientes se comprometan con contratos más largos desde el inicio.

---

### 3. Patrones de Facturación y Gasto se Correlacionan con la Evasión
**Observación:** Los clientes con cargos mensuales más altos tienden a tener tasas de evasión más elevadas en promedio. Por el contrario, los clientes con cargos totales acumulados más altos tienen menos probabilidad de abandonar.

**Recomendación:**
- Revisa la estrategia de precios y la percepción de valor.
- Analiza si los clientes con facturas mensuales elevadas sienten que no reciben suficiente valor.
- Explora estructuras de precios alternativas, servicios empaquetados o descuentos por fidelidad para clientes con alto gasto.
- Para clientes con cargos totales bajos (a menudo nuevos), enfócate en demostrar valor desde el inicio de su experiencia.

---

## 🧠 Acciones Estratégicas

- **Segmentación basada en datos:** Utiliza los segmentos de alto riesgo identificados para orientar estrategias de marketing, ventas y éxito del cliente.
- **Intervención dirigida:** Diseña e implementa iniciativas específicas de retención para grupos vulnerables.
- **Mejora de productos y servicios:** Usa los insights (por ejemplo, problemas con fibra óptica o pagos electrónicos) para impulsar mejoras en la oferta y entrega del servicio.
- **Gestión del ciclo de vida del cliente:** Desarrolla estrategias de compromiso específicas para cada etapa del ciclo de vida del cliente, con especial enfoque en el período inicial.
- **Monitoreo continuo:** Establece indicadores clave de rendimiento (KPIs) basados en estos hallazgos y monitorea regularmente las tasas de evasión por segmento, antigüedad y patrones de gasto para evaluar la efectividad de las estrategias implementadas.

---

Este análisis proporciona una hoja de ruta clara para enfocar los esfuerzos de retención donde tendrán mayor impacto. Al abordar la evasión dentro de los segmentos y períodos críticos identificados, el negocio puede mejorar el valor de vida del cliente y lograr un crecimiento sostenible.
